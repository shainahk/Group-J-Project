```python
df_stats = df.drop(['make', 'model', 'body', 'trim', 'vin', 'condition', 'odometer', 'color', 'interior', 'saledate'], axis= 1)
df_stats
```

```python
df_stats.dropna()
```

```python
df_stats.drop(df_stats.loc[df_stats['state']=="qc"].index, inplace=True)
df_stats.drop(df_stats.loc[df_stats['state']=="ab"].index, inplace=True)
df_stats.drop(df_stats.loc[df_stats['state']=="on"].index, inplace=True)
```

```python
state_dfs = {}
states = ['ca', 'tx', 'pa', 'mn', 'az', 'wi', 'tn', 'md', 'fl', 'ne', 'nj','nv', 'oh', 'mi', 'ga', 'va', 'sc', 'nc',
          'in', 'il', 'co', 'ut','mo', 'ny', 'ma', 'pr', 'or', 'la', 'wa', 'hi', 'ok', 'ms', 'nm', 'al', 'ns']

for state in states:
    state_df = df_stats.loc[df_stats['state'] == state]
    state_dfs[state] = state_df
```

```python
results = pd.DataFrame(columns=['state','dealership','lowest_price_mean'])

for state in states:
    df_stats2 = state_dfs[state]
    total_sellingprice = df_stats2['sellingprice'].groupby(df_stats2['seller'])
    df_stats2 = total_sellingprice.mean()
    lowest_deal = df_stats2.min()
    dealer = df_stats2.idxmin()
    results.loc[len(results.index)] = [state, dealer, lowest_deal]
    #print(state, dealer, lowest_deal)
```
